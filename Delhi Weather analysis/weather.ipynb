```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\VAIO\Desktop\Dataset 2\INDELHI.txt",header=None,sep="\t")
df.head()
```

```python
a=df[0].values
a
```

```python
month=[]
day=[]
year=[]
temp=[]
for i in a: 
    b=i.split()
    month.append(int(b[0]))
    day.append(int(b[1]))
    year.append(int(b[2]))
    temp.append(float(b[3]))
```

```python
df["Month"]=month
df["Year"]=year
df["day"]=day
df["temp"]=temp
```

```python
#converting temp from farhenite to celcious
df["temp"]=(df["temp"]-32)*(5/9)
df.head()
```
